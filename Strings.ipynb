### Finding Prime Numbers between 100 and 1000

This code block will identify and print all prime numbers within the specified range of 100 to 1000. It uses a helper function `is_prime` to determine if a number is prime by checking for divisibility only up to its square root, which optimizes the process.

```python
for i in range(100, 1001):
    for j in range(2, i // 2 + 1):
        if i % j == 0:
            break
    else:
        print(i, end=" ")
```

```python
password = input("Enter a password: ")
d=0
sc=0
lc=0
up=0
if len(password) > 7:
  for i in password:
    if i.isdigit():
      d=d+1
    elif i.isupper():
      up=up+1
    elif i.isalpha():
      lc=lc+1
    else:
      sc=sc+1
  if d>0 and lc>0 and up>0 and sc>0:
    print("Valid and STRONG Password")
  else:
    print("WEAK Password")
else:
  print("Password character must atleast >= 7")

```

```python
text = input("Enter the sentence: ")
cleaned = ""
for ch in text:
    if ch.isalpha():
        cleaned += ch

cleaned = cleaned.lower()

if cleaned == cleaned[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```
